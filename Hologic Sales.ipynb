# KNN to predict Item_Visibility values for 0 records

```python
train['Item_Identifier'].value_counts()
train.drop(['Item_Identifier'],axis=1,inplace=True)
```

```python
test=train.loc[train['Item_Visibility'] == 0]
train1=train.loc[train['Item_Visibility'] != 0]
```

```python
#train.iloc[  : train.columns != 'Item_Visibility' ]
X=train1[map(lambda x :x not in ['Item_Visibility'], list(train1.columns))].values
y=train1['Item_Visibility'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#import required packages
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rmse_val = [] #to store rmse values for different k
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(X_train, y_train)  #fit the model
    pred=model.predict(X_test) #make prediction on test set
    error = sqrt(mean_squared_error(y_test,pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('RMSE value for k= ' , K , 'is:', error)
```

```python
#plotting the rmse values against k values
curve = pd.DataFrame(rmse_val) #elbow curve 
curve.plot()
```

At K = 20 we get less RMSE value. Hence e take K = 20 and predict the missing Item_Visibility values

***Check skewness and kurtosis for these variables(Hypothesis test to get p-values and reduce the variable if p-value is  > 0.05, we can remove that variable from model since at p> 0.05, we'll always fail to reject null hypothesis***

```python
from scipy.stats import ttest_1samp
tset, pval = ttest_1samp(train['Item_Fat_Content'], 30)
print("Pvalues %Item_Fat_Content ",pval)

tset, pval = ttest_1samp(train['Item_MRP'], 30)
print("Pvalues %Item_MRP ",pval)

tset, pval = ttest_1samp(train['Item_Outlet_Sales'], 30)
print("Pvalues %Item_Outlet_Sales ",pval)

tset, pval = ttest_1samp(train['Item_Type'], 30)
print("Pvalues %Item_Type ",pval)

tset, pval = ttest_1samp(train['Item_Visibility'], 30)
print("Pvalues %Item_Visibility ",pval)

tset, pval = ttest_1samp(train['Item_Weight'], 30)
print("Pvalues %Item_Weight ",pval)

tset, pval = ttest_1samp(train['Outlet_Establishment_Year'], 30)
print("Pvalues %Outlet_Establishment_Year ",pval)

tset, pval = ttest_1samp(train['Outlet_Location_Type'], 30)
print("Pvalues %Outlet_Location_Type ",pval)

tset, pval = ttest_1samp(train['Outlet_Size'], 30)
print("Pvalues %Outlet_Size ",pval)

tset, pval = ttest_1samp(train['Outlet_Type'], 30)
print("Pvalues %Outlet_Type ",pval)
```

**Breusch-Pagan / Cook Weisberg Test** - This test is used to determine presence of **heteroskedasticity**. If you find p < 0.05, you reject the null hypothesis and infer that heteroskedasticity is present.

This test is performed on DV.
